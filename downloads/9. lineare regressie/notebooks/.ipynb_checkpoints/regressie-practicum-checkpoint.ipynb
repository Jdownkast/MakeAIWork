## Opdracht 1: Sea Level Change
### Global Average Absolute Sea Level Change, 1880-2014

https://www.kaggle.com/datasets/somesh24/sea-level-change

### Doel

- Inzicht in zeespiegelstijging
- Model maken dat stijging kan voorspellen

### Stap 1: Data inladen

```python
# Use pandas to load 'sea_levels_2015.csv'
import pandas as pd
import numpy as np
import scipy as cp
import datetime as dtime

sea_file = "../data/sea_levels_2015.csv"

df_sea = pd.read_csv(sea_file)
print(df_sea)

# To check if this set is complete, use:

df_sea.isnull().sum()
```

### Stap 2: Data bekijken

```python
# Plot the sea level data
import matplotlib.pyplot as plt

sea_plot = plt.subplots()

time_date = [dtime.datetime.strptime(date, '%Y-%m-%d') for date in df_sea['Time']]
time = np.array([(date - time_date[0]).days for date in time_date])
gmsl = df_sea['GMSL']

print(gmsl)
plt.plot(time, gmsl)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
plt.clf()
```

### Stap 3: Tends zoeken

- Zijn er trends zichtbaar in de data?
- Zo ja, welke zijn dat?
- Zoom desnoods wat in; neem als range 1 jaar

Bouw op basis van deze inzichten een model.

```python
# def model(a, b, c, ...):
def sea_level(time, const, coeff):
    return const + time * coeff

plt.plot(time, gmsl)
plt.gca().axes.get_xaxis().set_visible(False)

# FOR ZOOM
# plt.axis([time[1550], time[1600], 0, 100])
# plt.xticks(np.arange(time[1550], time[1600], 182), rotation = 90)


# Fit your model

popt, pcov = cp.optimize.curve_fit(sea_level, time, gmsl)

# Plot your fit

plt.plot(time, gmsl, 'b')
plt.plot(time, sea_level(time, *popt), 'r--')
plt.show()
```

### Stap 4: Voorspellen

Kun je op basis van je model voorspellen wat de gemiddelde zeespiegel is in 2100?

```python
# Extrapolate to 2100 with your model

date_str = "2100-01-01"
date_value = (dtime.datetime.strptime(date_str, '%Y-%m-%d') - time_date[0]).days


def sea_predict(dateval, constant, coefficient):
    return constant + dateval * coefficient

prediction = sea_predict(date_value, popt[0], popt[1])
print(prediction)


# Optional: can you plot the future up to 2100?
```

Is je uitkomst verontrustend?
