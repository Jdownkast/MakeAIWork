```python
import pandas as pd
from IPython.display import display, HTML
import matplotlib.pyplot as plt
```

```python
# Load data
```

```python
filename = "data/chronicgdaytotpop.xlsx"
df_eu_food = pd.read_excel(filename, header=[2])

# def show_df(df_name):
#     # return display(HTML(df_name.to_html()))
#     return display(df_name)
```

```python
# Manipulate data:

# 1. Choose category
# 2. Check if all countries are represented, else find different category
# 3. Compromise subset with only the new category
# 4. Repeat for more data

df_eu_food['Total'] = df_eu_food['Nr Subjetcs'] * df_eu_food['Mean']

cat_all = pd.unique(df_eu_food['Foodex L1'])
land_all = pd.unique(df_eu_food['Country'])
# print([x for x in land_all])  

df_eu_new = pd.DataFrame()
df_eu_new.index = land_all

for cat in cat_all:
    df_eu_cat = df_eu_food[df_eu_food['Foodex L1'].str.startswith(cat)]

    df_eu_cat_sub = df_eu_cat.groupby('Country')[['Nr Subjetcs', 'Total']].sum()
    df_eu_cat_sub[f'Mean {cat}'] = df_eu_cat_sub['Total'] / df_eu_cat_sub['Nr Subjetcs'] 
    # df_eu_cat_sub_sorted = df_eu_cat_sub.sort_values([f'Mean {cat}'], ascending=False)
    
    df_eu_new[f'Mean {cat}'] = df_eu_cat_sub[f'Mean {cat}']

# display(df_eu_food)
# display(df_eu_cat)
# display(df_eu_cat_sub)
# display(df_eu_cat_sub_sorted)

display(df_eu_new)

# df_subset = df_eu_food[['Mean','Total']]
# display(df_subset)

# df_eu_new.to_csv("data/EU_food_consumption.csv")
```
